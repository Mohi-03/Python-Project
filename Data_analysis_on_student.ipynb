```python
# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

```python
# Fill missing values for simplicity
df.ffill(inplace=True) # forward fill
df.bfill(inplace=True) # backword fill
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Visualizations
df['guardian'].value_counts().plot(kind='barh')
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='Fjob',data=df ,edgecolor='black')
plt.title("Count of Types of Fathers job ")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(x='Mjob',data=df ,edgecolor='black')
plt.title('Count of Types of Mothers job')
plt.show()
```

```python
# Histogram of Student Ages
plt.subplot(2, 2, 1)
plt.hist(df['age'], bins=10, edgecolor='black')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
```

```python
#Bar Plot of Gender Distribution

plt.subplot(2, 2, 2)
sns.countplot(x='sex', data=df , edgecolor='black')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True)
plt.title('Distribution of Student Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Box **plot**

```python
 # Box Plot for Final Grades (G3)

plt.subplot(2, 2, 3)
sns.boxplot(x=df['G3'])
plt.title('Distribution of Final Grades')
plt.xlabel('Final Grade')
```

```python
sns.boxplot(df,x='age',y='absences')
```

pie **chart**

```python
# Count occurrences of a specific categorical column (e.g., 'sex')
category_counts = df['sex'].value_counts()
```

```python
# Plot pie chart

plt.figure(figsize=(4, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140 )
plt.title('Distribution of Gender')
plt.show()
```

Scatter **plot**

```python
# Scatter plot

plt.scatter(df['traveltime'],df['studytime'])
```

```python
# clean data delete missing values in 'studytime' simplicity

df_cleaned = df.dropna(subset=['studytime', 'G3'])

# Prepare data for plotting
x = df_cleaned['studytime']
y = df_cleaned['G3']
```

```python
 # scatter plot
plt.figure(figsize=(5, 3))
plt.scatter(x, y, color='blue', marker='o', alpha=0.7)
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time (hours per week)')
plt.ylabel('Final Grade (G3)')
plt.grid(True)
plt.show()
```
